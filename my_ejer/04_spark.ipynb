```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark import SparkContext
from pyspark.sql import SQLContext
import pandas as pd
```

```python
spark = SparkSession.builder.getOrCreate()  # creamos una instancia de la clase SparkSession
sc = spark.sparkContext # creamos una instancia de la clase SparkContext
```

```python
sqlContext = SQLContext(sc) # creamos una instancia de la clase SQLContext
df = sqlContext.read.csv('./dataset/bigmac.csv', header=True, inferSchema=True) # leemos el archivo csv
rdd = df.rdd    # convertimos el dataframe a un RDD
```

```python
df.show(5)  # mostramos los primeros 5 registros del dataframe
```

```python
rdd.count() # contamos el número de filas del RDD
```

```python
rdd.take(5) # mostramos las primeras 5 filas del RDD
```

```python
rdd.first() # mostramos la primera fila del RDD
```

```python
rdd.filter(lambda x: x.name == "Argentina").count() # contamos el número de filas que cumplen la condición
```

```python
# usamos collect para obtener una lista con los valores de la columna name
rdd.map(lambda x: x.name).collect()

```
